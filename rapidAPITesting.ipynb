```python
USA_LOCATION_TERMS = {
    "usa", "worldwide", "americas", "us only",
    "united states", "north america", "us timezones",
    "usa timezones", "remote"
}

def _is_usa_eligible(location_str: str) -> bool:
    """
    Returns True if the job is open to US-based candidates.
    Remotive location is a free text string — we check for
    known USA-inclusive terms. If no location listed we allow
    it through since no restriction means open.
    """
    if not location_str.strip():
        return True
    loc_lower = location_str.lower()
    return any(term in loc_lower for term in USA_LOCATION_TERMS)

# Test against real strings from our earlier output
test_locations = [
    "Europe, European timezones",      # should fail
    "USA timezones, European timezones",# should pass
    "Worldwide",                        # should pass
    "USA",                              # should pass
    "Brazil, Colombia, Philippines",    # should fail
    "Americas, Europe, Israel",         # should pass
    "USA, Canada, USA timezones",       # should pass
    "Germany",                          # should fail
    "LATAM, Asia",                      # should fail
    "",                                 # should pass — no restriction
]

print("Location filter test:\n")
for loc in test_locations:
    result = _is_usa_eligible(loc)
    status = "✅ pass" if result else "✗  fail"
    print(f"  {status} | '{loc}'")
```

```python
from bs4 import BeautifulSoup
from datetime import datetime

# Maps our UI dropdown values to Remotive category slugs
# Two slugs per category — we run both and combine
REMOTIVE_CATEGORY_MAP = {
    "Software Engineering": ["software-development", "artificial-intelligence"],
    "Data and Analytics":   ["data",                 "artificial-intelligence"],
    "Computer and IT":      ["information-technology","devops"],
}

def _strip_html(html: str) -> str:
    if not html:
        return ""
    return BeautifulSoup(html, "html.parser").get_text(separator=" ", strip=True)


def search_jobs_remotive(
    category: str,
    job_titles: str = "",
    limit_per_slug: int = 20,
) -> list[dict]:
    """
    Fetch jobs from Remotive across all slugs mapped to the given category.
    Filters to USA-eligible jobs only.
    Returns normalized job dicts matching save_jobs() schema.
    No level filter — LLM handles seniority filtering downstream.
    """
    slugs = REMOTIVE_CATEGORY_MAP.get(category, ["software-development"])

    raw_jobs = []
    for slug in slugs:
        print(f"[search_jobs_remotive] Fetching slug='{slug}'")
        try:
            response = requests.get(REMOTIVE_BASE_URL, params={
                "category": slug,
                "limit":    limit_per_slug,
            })
            response.raise_for_status()
            batch = response.json().get("jobs", [])
            print(f"[search_jobs_remotive] '{slug}' returned {len(batch)} raw jobs")
            raw_jobs.extend(batch)
        except Exception as e:
            print(f"[search_jobs_remotive] Failed for slug '{slug}': {e}")
            continue

    # Deduplicate by ID across slugs — same job can appear in multiple categories
    seen_ids = set()
    unique   = []
    for job in raw_jobs:
        jid = job.get("id")
        if jid and jid not in seen_ids:
            seen_ids.add(jid)
            unique.append(job)

    print(f"[search_jobs_remotive] {len(unique)} unique jobs after slug dedup")

    # Normalize + location filter
    jobs = []
    for job in unique:
        url = job.get("url", "")
        if not url:
            continue

        description = _strip_html(job.get("description", ""))
        if not description:
            continue

        location_raw = job.get("candidate_required_location", "")
        if not _is_usa_eligible(location_raw):
            continue

        # Parse salary string into min/max floats if possible
        salary_str = job.get("salary", "") or ""
        salary_min, salary_max = _parse_salary(salary_str)

        jobs.append({
            "title":       job.get("title", ""),
            "company":     job.get("company_name", ""),
            "location":    location_raw,
            "salary_min":  salary_min,
            "salary_max":  salary_max,
            "description": description,
            "url":         url,
            "created":     job.get("publication_date", datetime.utcnow().isoformat()),
        })

    print(f"[search_jobs_remotive] {len(jobs)} jobs after USA location filter")
    return jobs
```

```python
import re

def _parse_salary(salary_str: str) -> tuple:
    """
    Extract salary_min and salary_max from Remotive's free text salary field.
    Handles formats like '$80k - $120k', '$12k - $19k (35 hrs/week)'
    Returns (None, None) if parsing fails.
    """
    if not salary_str:
        return None, None
    try:
        # Find all numbers, handle 'k' multiplier
        nums = re.findall(r"\$?([\d,]+\.?\d*)\s*k?", salary_str.lower())
        cleaned = []
        for n in nums:
            val = float(n.replace(",", ""))
            # If original string had 'k' after this number, multiply
            if "k" in salary_str.lower():
                val *= 1000
            cleaned.append(val)
        if len(cleaned) >= 2:
            return cleaned[0], cleaned[1]
        elif len(cleaned) == 1:
            return cleaned[0], None
    except Exception:
        pass
    return None, None

# Test it
test_salaries = [
    "$80k - $120k",
    "$12k - $19k (35 hrs/week, freelance)",
    "$150,000 - $200,000",
    "",
    None,
]
for s in test_salaries:
    mn, mx = _parse_salary(s or "")
    print(f"  '{s}' → min={mn} max={mx}")
```
